# Генерация данных

```python
import pandas as pd
import numpy as np

n = 500

df=pd.DataFrame({'код респондента': np.random.randint(1,np.int(0.33*n),size=n), 
                 'дата опроса': np.random.choice(['2020-01-31','2021-01-31','2022-02-03','2019-06-06'],size=n), 
                 'пол': np.random.choice(['муж','жен'],size=n,p=(0.35,0.65)), 
                 'образование': np.random.choice(['высшее','незаконченное высшее', 'среднее','среднее специальное'],p=(0.1,0.1,0.4,0.4),size=n), 
                 'социальный статус': np.random.choice(['холост/незамужем','женат/замужем', 'вдовец/вдова','разведен/разведена'],p=(0.45,0.05,0.05,0.45),size=n), 
                 'возраст': np.rint(np.random.normal(45,15,size=n)), 
                 'рост': list(np.rint(np.random.normal(160,20,size=int(n*0.9)))) + list(np.rint(np.random.normal(180,20,size=n-int(n*0.9))))})

df['вес'] = df['рост']-(100+(df['рост']-100)/20)+np.rint(np.random.normal(10,2,size=n))
```

```python
def scatter_hist(x, y, ax, ax_histx, ax_histy):
    # no labels
    ax_histx.tick_params(axis="x", labelbottom=False)
    ax_histy.tick_params(axis="y", labelleft=False)

    # the scatter plot:
    ax.scatter(x, y)

    # now determine nice limits by hand:
    binwidth = 0.25
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth) + 1) * binwidth

    bins = np.arange(-lim, lim + binwidth, binwidth)
    ax_histx.hist(x, bins=bins)
    ax_histy.hist(y, bins=bins, orientation='horizontal')
```

```python
fig = plt.figure(figsize=(16, 10))

gs = fig.add_gridspec(4, 4, hspace=0.5, wspace=0.2)

ax_scatter = fig.add_subplot(gs[:-1, :-1])
ax_right = fig.add_subplot(gs[:-1, -1], xticklabels=[], yticklabels=[])
ax_bottom = fig.add_subplot(gs[-1, 0:-1], xticklabels=[], yticklabels=[])

ax_scatter.scatter(df['вес'], df['рост'])

ax_bottom.hist(df['вес'], 40, orientation='vertical')
ax_bottom.invert_yaxis()

ax_right.hist(df['рост'], 40, orientation='horizontal')

ax_scatter.set(title='Диаграмма рассеяния вес/рост с построчными гистограммами', xlabel='вес', ylabel='рост')

for item in ([ax_scatter.xaxis.label, ax_scatter.yaxis.label] + ax_scatter.get_xticklabels() + ax_scatter.get_yticklabels()):
    item.set_fontsize(14)

plt.show()
```
